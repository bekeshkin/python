### Колонка "Cuisine Style"

```python
print(type(df['Cuisine Style'][0]))
df['Cuisine Style'][0]
```

Содержит список кухни, блюда которых подаются в ресторане. Данные хранятся ввиде списка, но преобразованы в строку

* его можно преобразовать в набор дополнительных признаков, появляется ли та или иная кухня в данном ресторане
* можно добавить признак Cuisine Style Info, в котором 1 - это присутствует информация о кухне, 0 - отсутствует
* можно добавить признак Cuisine Count - разнообразие кухни

```python
## заменим отсутсвие данных на "['No data']" чтобы в дальнейшем такие строки можно было бы обработать
df['Cuisine Style'] = df['Cuisine Style'].fillna("['No data']")
df['Cuisine Style'] = df['Cuisine Style'].apply(lambda x: literal_eval(str(x)))
```

```python
## составим список кухонь
cs = set() 
for x in df['Cuisine Style']:
    cs.update(x)
```

```python
## добавим дополнительные признаки в данные вида "Название кухни": 
## 1 - блюда данной кухни подаются в ресторане
## 0 - не подаются
def find_item(cell):
    if item in cell:
        return 1
    return 0

for item in cs:
    df[item] = df['Cuisine Style'].apply(find_item)
```

```python
## посчитаем количество кухонь
def count_cs(lst):
    if len(lst) == 0:
        return 1
    return len(lst)
df['Cuisine Count'] = df['Cuisine Style'].apply(count_cs)
```

```python
## определим наличие данных о кухне
df['Cuisine Style Info'] = df['No data'].apply(lambda x: 0 if x == 1 else 1)
```

```python
## удалим 'Cuisine Count' и 'No data' из них больше информации не извлечь
df = df.drop(['Cuisine Style', 'No data'], axis = 1)
```

### Колонка "Price Range"

```python
df['Price Range'].value_counts()
```

колонка содержит категориальный признак ценовой категории ресторана, опять же только у чуть больше половины есть данные

* необходимо присвоить каждой категории (в том числе для NaN) числовой признак (0 - NaN, 1 - низкий ценовой сегмент, 2 - средний, 3 - высокий ценовой сегмент
* также как и для списка кухни можно добавить отдельную колонку по наличию информации Price Range Info

```python
def setPriceRange(value):
    if value == '$':
        return 1 ## низкая ценовая категория
    if value == '$$ - $$$':
        return 2 ## средняя ценовая категория
    if value == '$$$$':
        return 3 ## наибольшая ценовая категория
    return 0 ## нет данных
df['Price Range'] = df['Price Range'].apply(lambda x: setPriceRange(x))
```

```python
df['Price Range Info'] = df['Price Range'].apply(lambda x: 0 if x == 0 else 1)
```

### Колонка "Reviews"

```python
print(df['Reviews'][0])
type(df['Reviews'][0])
```

Содержит информацию о двух отзывах, размещенных на сайте, как видно представляет собой двумерный массив, где в первом списке находятся отзывы, а во втором списке - даты этих отзывов. Но представлен ввиде строки

```python
## преобразуем поле Review чтобы в дальнейшем его можно было легче обработать
from datetime import datetime

def reviewsConveter(value):
    if isinstance(value, str):
        ## преобразования, чтобы literal_eval отработал корректно
        value = value.replace("nan]", "'']") 
        value = value.replace("[nan", "[''")
        ## преобразование строки в список
        value = literal_eval(value)
    
    ## если список преобразован, преобразуем список дат из строкового представления в тип date
    if isinstance(value, list):
        dates = value[1]
        value[1] = [datetime.strptime(date, '%m/%d/%Y').date() for date in dates]
    return value

df['Reviews'] = df['Reviews'].fillna("[[],[]]") ## заполним отсутствующие данные
df['Reviews'] = df['Reviews'].apply(lambda x: reviewsConveter(x))
```

Добавим 4 колонки
* First Review - первый отзыв
* First Review Date - дата первого отзыва
* Second Review - второй отзыв
* Second Review Date - дата второго отзыва

```python
df['Reviews'][0][0]
```

```python
def getReview(review, row, col):
    try:
        return review[row][col];
    except:
        return None;
```

```python
df['First Review'] = df['Reviews'].apply(lambda x: getReview(x, 0, 0))
df['First Review Date'] = df['Reviews'].apply(lambda x: getReview(x, 1, 0))
df['Second Review'] = df['Reviews'].apply(lambda x: getReview(x, 0, 1))
df['Second Review Date'] = df['Reviews'].apply(lambda x: getReview(x, 1, 1))
```

Видим, что отзывов не всегда 2, а иногда они отсутствуют.
Можно добавить следующие колонки
* Last Review Date - дата последнего отзыва
* Days After Review - количество дней с момента публикации последнего отзыва
* Review Count - количество отзывов на сайте

```python
def lastReviewDate(x):
    if (x['First Review Date'] != None) and (x['Second Review Date'] == None):
        return x['First Review Date']
    if (x['First Review Date'] == None) and (x['Second Review Date'] != None):
        return x['Second Review Date']
    if (x['First Review Date'] != None) and (x['Second Review Date'] != None):
        return max([x['First Review Date'], x['Second Review Date']])
    return None
```

```python
df['Last Review Date'] = df.apply(lambda x: lastReviewDate(x), axis=1)
```

```python
## количество дней после даты, если дата отсутствует, то None
def daysAfter(x):
    today = datetime.now().date()
    if (x != None):
        return (today-x).days
    return None;
```

```python
df['Days After Review'] = df['Last Review Date'].apply(lambda x: daysAfter(x))
```

```python
## функция подсчета колчества отзывов
def countReviewOnSite(x):
    if (x['First Review Date'] == None) and (x['Second Review Date'] == None):
        return 0
    if (x['First Review Date'] != None) and (x['Second Review Date'] == None):
        return 1
    return 2

df['Count Review'] = df.apply(lambda x: countReviewOnSite(x), axis=1)
```

### Колонка 'Number of Reviews'

```python
## заполним недостающие данные в данной колонке в зависимости от наличия отзывов в колонке Count Review
df['Number of Reviews'] = df['Number of Reviews'].fillna(df['Count Review'])
```

# Разбиваем датафрейм на части, необходимые для обучения и тестирования модели

```python
# Х - данные с информацией о ресторанах, у - целевая переменная (рейтинги ресторанов)
X = df.drop(['Restaurant_id', 'Rating'], axis = 1)
y = df['Rating']
```

```python
# Загружаем специальный инструмент для разбивки:
from sklearn.model_selection import train_test_split
```

```python
# Наборы данных с меткой "train" будут использоваться для обучения модели, "test" - для тестирования.
# Для тестирования мы будем использовать 25% от исходного датасета.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

# Создаём, обучаем и тестируем модель

```python
# Импортируем необходимые библиотеки:
from sklearn.ensemble import RandomForestRegressor # инструмент для создания и обучения модели
from sklearn import metrics # инструменты для оценки точности модели
```

```python
# Создаём модель
regr = RandomForestRegressor(n_estimators=100)

# Обучаем модель на тестовом наборе данных
regr.fit(X_train, y_train)

# Используем обученную модель для предсказания рейтинга ресторанов в тестовой выборке.
# Предсказанные значения записываем в переменную y_pred
y_pred = regr.predict(X_test)
```

```python
# Сравниваем предсказанные значения (y_pred) с реальными (y_test), и смотрим насколько они в среднем отличаются
# Метрика называется Mean Absolute Error (MAE) и показывает среднее отклонение предсказанных значений от фактических.
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
```
